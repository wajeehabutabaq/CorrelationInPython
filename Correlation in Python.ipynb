```python
# Importing

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Data reader from csv

df = pd.read_csv(r"C:\Users\Wajeeh Abu Tabaq\Documents\movies.csv\movies2.csv")
```

```python
df.head()
```

```python
# Missing Data check - I cleaned it manually in excel beforehand, but just to check

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Data type check 

df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['Month'] = df['released'].astype(str).str[:3]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# For non numeric data something clear that should have a relation is the company
#Numerize

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correl_matrix = df_numerized.corr()
sns.heatmap(correl_matrix, annot =True)

plt.title('Matrix')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show
```

```python
#Less color here

df_numerized.corr()
```

```python
correl_mat = df_numerized.corr()

corr_pairs = correl_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Lets do more than 0.5

high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
